### Перепишем функцию на зыке питона в цикле

```python
def f(x):
    b = []
    for i in range(len(x)):
        b.append (sin(x[i] / 5) * exp(x[i] / 10) + 5 * exp(-x[i] / 2))
    return b
```

###### Проверим правильность функции

```python
x = [1,15]
b = f(x)
plt.plot(b)
plt.show()
```
