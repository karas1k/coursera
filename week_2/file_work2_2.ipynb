```python
from math import *
from numpy import *
from matplotlib import pylab as plt
```

```python
def f(x):
    return sin(x/5) * exp(x/10) + 5*exp(-1*x/2)
```

```python
def Y(x, W):
    x_copy = x.copy()
    x_copy = x_copy * 0
    i = 0
    for w in W:
        x_copy += w * (x ** i)
        i += 1
    return x_copy
    W
```

```python
a = array([[1, 1], [1, 15]])
b = array([f(1), f(15)])
W1 = linalg.solve(a, b)
W1
```

```python
x = arange(0, 16, 0.1)
y1 = f(x)
y2 = Y(x, W1)
plt.plot(x, y1, x, y2)
plt.show()
```

```python
a = array([[1, 1, 1], [1, 8, 8 ** 2], [1, 15, 15 ** 2]])
b = array([f(1), f(8), f(15)])
W = linalg.solve(a, b)
W
```

```python
x = arange(0, 16, 0.1)
y3 = f(x)
y4 = Y(x, W)
plt.plot(x, y3, x, y4)
plt.show()
```
